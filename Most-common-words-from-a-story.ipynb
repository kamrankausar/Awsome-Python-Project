```python
# Load the file, and the encoding 
file = open('PrideandPrejudice.txt', encoding= 'utf8')
a = file.read()
```

```python
# Handling Stop words
stopwords = set(line.strip() for line in open('stopwords.txt', encoding= 'utf8'))
stopwords.union(set(['mr', 'mrs', 'one', 'two', 'said']))
```

```python
# Instantiate a dictonary, and for every word in the file, 
# Add to the dictonary if it doesn't exist. If it does, increase the count
wordcount = {}
```

```python
# To eliminate the Duplicate, remember to split by punctuation, and use case delimiters

for word in a.lower().split():
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("\"","")
    word = word.replace("!","")
    word = word.replace("â€œ","")
    word = word.replace("â€˜","")
    word = word.replace("*","")
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
```

```python
# Print most common word
n_print = int(input("How many most common words to print: "))
print("\nOK. The {} most common words are as follows\n".format(n_print))
```

```python
word_counter = collections.Counter(wordcount)
for word, count in word_counter.most_common(n_print):
    print(word, ": ", count)
```

```python
# At last Close the file
file.close()
```

```python
# Create a data frame of the most common words 
# Draw a bar chart
lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word', 'Count'])
df.plot.bar(x='Word',y='Count')
```
